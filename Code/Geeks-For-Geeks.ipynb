#         

### BeautifulSoup parsing HTML

```python
soup = be(req.content, 'html.parser')
```

```python
#Getting the title of the page
soup.title
```

```python
#Getting the name of the tag
soup.title.name
```

```python
#Getting the name of parent tag
soup.title.parent.name
```

```python
#Getting the name of child tag
soup.title.child
```

#       

### Finding Elements

```python
#entry-content, <p>
# Go to the first div that his class is entry-content
s = soup.find('div', class_='entry-content')
print(s)

# After that find all <p> in this div
content = s.find_all('p')
print('======================================\n\n\n\n\n')
print(content)
```

#      

### Finding element by ID

```python
s = soup.find('div', id='main')
print(s)
print('============================================\n\n\n\n\n\n\n')

content = s.find_all('ul', class_='leftBarList')
print(content)
```

#      

### Extract text from tags

```python
#Removing the tags from the content
s = soup.find('div', class_='entry-content')
content = s.find_all('p')

for i in content:
    print(i.text)
```

#      

```python
s = soup.find('div', id='main')
content = s.find_all('li')

for i in content:
    print(i.text)
```

#                     

### Extract the links

```python
for i in soup.find_all('a'):
    print(i.get('href'))
```

##            

### Extract Images

```python
images = soup.find_all('img')
for i in images:
    print(i.get('src'))
```

```python
image_dict = {}

images = soup.select('img')
for image in images:
    src = image.get('src')
    alt = image.get('alt')
    image_dict[alt] = src

for key, value in image_dict.items():
    print(f'{key} : {value}', end='\n\n')
```
