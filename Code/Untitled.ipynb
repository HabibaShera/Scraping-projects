```python
import requests
from bs4 import BeautifulSoup as bs

import csv
import pandas as pd
```

```python
# Access the website
url = 'https://eg1lib.org/s/programming%20concepts/?type=phrase'
request = requests.get(url)
request
```

```python
soup = bs(request.content,"html.parser")
```

```python
containers = soup.find_all('div', class_="resItemBox resItemBoxBooks exactMatch")
len(containers)
```

```python
title = containers[30].find_all('h3', itemprop="name")
title[0].text.strip()
```

```python
publisher = containers[40].find_all('a', title="Publisher")
publisher[0].text
```

```python
author = containers[4].find_all('div', class_="authors")
author[0].text
```

```python
year = containers[49].find_all('div', class_="property_value")
int(year[0].text)
```

```python
for container in containers:
    title = container.find_all('h3', itemprop="name")
    title = title[0].text.strip()
    
    try:
        publisher = container.find_all('a', title="Publisher")
        publisher = publisher[0].text
    except:
        publisher = 'nan'
    
    author = containers[4].find_all('div', class_="authors")
    author = author[0].text
    
    year = containers[49].find_all('div', class_="property_value")
    year = int(year[0].text)
    
    print(title + '\n' + publisher + '\n' + author + '\n' + str(year) + '\n')
```

```python
with open('zz.csv', 'w', encoding='utf-8') as f:
    obj = csv.writer(f)
    obj.writerow(['tile', 'publisher', 'author', 'year'])
    
    for container in containers:
        title = container.find_all('h3', itemprop="name")
        title = title[0].text.strip()

        try:
            publisher = container.find_all('a', title="Publisher")
            publisher = publisher[0].text
        except:
            publisher = 'nan'

        author = containers[4].find_all('div', class_="authors")
        author = author[0].text

        year = containers[49].find_all('div', class_="property_value")
        year = int(year[0].text)

        #print(title + '\n' + publisher + '\n' + author + '\n' + str(year) + '\n')
        obj.writerow([title, publisher, author , year])
        
    print('Done.....')
```
