```python
import requests
from bs4 import BeautifulSoup as bs

import shutil
import os
```

```python
url = 'https://eg1lib.org/category/173/Computer-Science'
req = requests.get(url)
req
```

```python
soup = bs(req.content, 'html.parser')
```

```python
containers = soup.find_all('div', {'class':'resItemBox resItemBoxBooks exactMatch'})
len(containers)
```

```python
img = soup.find_all('img', {'class':'cover lazy'})
len(img)
```

```python
title = soup.find_all('h3', itemprop="name")
len(title)
```

```python
titles = []
img_links = []

for i in range(50):
    titles.append(title[i].text.strip())
    img_links.append(img[i]['data-src'])

library = dict(zip(titles, img_links))

for key, value in library.items():
    print(f'{key} : {value}', end='\n\n')
```

#    

The `copyfileobj()` method to write your image as the file name, builds the file locally in binary-write mode and saves it locally with shutil. Although it's not necessary, it's good to check if the image was retrieved successfully using Request’s Status Code in a conditional statement.

```python
for i in range(len(img_links)):
    request = requests.get(img_links[i], stream = True)
    if request.status_code==200:
        name = titles[i]
        with open( name+'.jpg', 'wb') as f:
            shutil.copyfileobj(request.raw, f)
            print(name + ' done.....')
```
